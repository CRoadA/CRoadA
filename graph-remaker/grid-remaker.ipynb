```python
# could be replaced by some geopandas stuff?
@dataclass
class Point:
    x: float
    y: float

@dataclass
class Street:
    id: int
    points: list[Point]

@dataclass
class Crossroad:
    streets: list[int] # ids

class FoundStreets:

    def __init__(self):
        self.streets: list[Street] = []
        self.disruptions: dict[Street, list[Point]] = {} # list of problematic points by their strets
        self.crossroads: list[Crossroad]


    
```

```python
def find_first_non_checked_road(grid: np.ndarray[(Any, Any), bool], checked: np.ndarray[(Any, Any), bool]) -> tuple[bool, tuple[int, int]]:
    for (x, y), value in np.ndenumerate(checked):
        if value == 0:
            if grid[x, y] == True:
                return True, (x, y), value
            checked[x, y] = 1
    return False, None
```

```python
def identify_streets(grid_part: np.ndarray[(Any, Any, Any), Any]) -> FoundStreets:
    checked = np.zeros(grid_part, dtype=bool)
    is_found, (row, col) = find_first_non_checked_road(grid_part)
    results = FoundStreets()
    while is_found:
        # go for it and identify street
        is_found, (row, col) = find_first_non_checked_road(grid_part)
```

```python
def divide_into_parts(grid: np.ndarray[(Any, Any, Any), Any], part_size: int = 200):
    width, height = grid.shape
    x_parts_number = math.ceil(width / part_size)
    y_parts_number = math.ceil(height / part_size)

    results = []
    for row in range(x_parts_number):
        # lower/upper in the sense of indices values
        # parts overlap by 1 row
        y_lower = max(0, row * part_size - 1)
        y_upper = min(0, (row + 1) * part_size + 1)
        for col in range(y_parts_number):
            # parts overlap by 1 column
            x_lower = max(0, col * part_size - 1)
            x_upper = min(width, (col + 1) * part_size + 1)
            results.push(identify_streets(grid[y_lower:y_upper, x_lower:x_upper]))
    # and so on...
```
