```python
import os
import numpy as np
from scraper.data_loader import DataLoader
from grid_manager import GridManager 
```

```python
city_test_name = "Kraków"
filename = "krakow_test.grid"
data_dir = "grids"

```

```python
file_path = os.path.join(data_dir, filename)
if os.path.exists(file_path):
    os.remove(file_path)
    print(f"Usunięto stary plik: {filename}")
```

```python
loader = DataLoader(grid_density=1.0, segment_h=1000, segment_w=1000, data_dir=data_dir)
```

```python
grid_manager = loader.load_city_grid(city_test_name, filename)
```

```python
loader.add_elevation_to_grid(grid_manager)
```

```python
segment = grid_manager.read_segment(0, 0)
```

```python
h, w, layers = segment.shape
print(f"Wymiary pobranego segmentu: {h}x{w}, warstwy: {layers}")
```

```python
mid_y, mid_x = h // 2, w // 2
road_val = segment[mid_y, mid_x, 0]    # Warstwa 0: Drogi
altitude_val = segment[mid_y, mid_x, 1] # Warstwa 1: Wysokość
```

```python
print(f"\nPrzykładowy punkt [{mid_y}, {mid_x}]:")
print(f"  -> Czy jest drogą (0/1): {road_val}")
print(f"  -> Wysokość (Z): {altitude_val}")
```

```python
min_h = np.min(segment[:, :, 1])
max_h = np.max(segment[:, :, 1])
mean_h = np.mean(segment[:, :, 1])
```

```python
print(f"\nStatystyki wysokości w segmencie:")
print(f"  -> Min: {min_h}")
print(f"  -> Max: {max_h}")
print(f"  -> Średnia: {mean_h}")
```
