```python
import sys
import os
import osmnx as ox
from rasterizer import Rasterizer
import math

project_path = os.path.join(os.getcwd(), "..")
sys.path.append(project_path)

from grid_builder import GridBuilder

builder = GridBuilder()
gdf_edges = builder.get_city_roads("Kraków, Polska")
min_x, min_y, max_x, max_y = gdf_edges.total_bounds
columns_number = max_x - min_x
rows_number = max_y - min_y
segment_w = 10000
segment_h = 14000
grid_density = 1

segment_rows = math.ceil((rows_number)/(segment_h * grid_density))
segment_cols = math.ceil((columns_number)/(segment_w * grid_density))
print(f"Height: {int(rows_number)}, Width: {int(columns_number)}, rows: {segment_rows}, cols: {segment_cols}")

rasterizer = Rasterizer()
# for i in range(segment_rows):
#     for j in range(segment_cols):
grid = rasterizer.rasterize_segment_from_coordinates(gdf_edges=gdf_edges, min_x=min_x, max_x=min_x+10000, min_y=min_y, max_y=min_y+14000, pixel_size=grid_density)
        # print(f"Segment: {i}, {j}")
        # print(f"Grid segment shape: {grid.shape}")
```
